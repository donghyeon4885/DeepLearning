# 다중선형회귀

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x1= np.array([2,4,6,8])
x2 = np.array([0,4,2,3])
y= np.array([81,93,91,97])
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter3D(x1,x2,y)
plt.show()
```

```python
a1, a2, b = (0,0,0)     # 기울기의 값을 초기화
lr = 0.01
epochs =2001
n = len(x1)
y_pred= a1*x1 + a2 * x2 + b
```

# 경사하강법을 시작

```python
for i in range(epochs):
    y_pred = a1*x1 +a2 * x2+b  # 예측값 구함
    error = y- y_pred  # 실제값과 예측값의 차이
    a1 = a1- lr *( (2/n) *sum(-x1 *(error)))
    a2 = a2- lr *( (2/n) *sum(-x2 *(error)))
    b = b-lr * ((2/n) *sum(-(error)))
                
    if i % 100 == 0:
        print('epochs : {}, 기울기1 : {:3.2f}, 기울기2 : {:3.2f}, 절편 :{}'.format(i,a1,a2,b))
                
# 실제 예측된 최종 점수 출력
y_pred = a1*x1+ a2 *x2+b
print('실제 점수 : ', y)
print('예측 점수 : ',y_pred)

# 공주 시간이 5시간이고 과외 횟수가 3인 학생의 점수 예측
print('예측점수 : ', a1*5+a2*3+b)
```

# 탠서플로우와 케라스를 활용한 선형 회귀 분석

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense       # lyaer 을 추가하기 위해서 Dense 클래스 사용
model = Sequential()                             # 모델 생성 -> 모델 초기화
model.add(Dense(1, input_dim=1,activation='linear'))       # layer 추가
model.summary()
model.compile(optimizer='sgd', loss='mse')            # 모델을 실행할 파라미터 설정
model.fit(x1,y, epochs=2000)                  # 모델 학습
```

```python
plt.scatter(x1,y)
plt.plot(x1,model.predict(x1),'r')
plt.show()
```

```python
# 공부시간이 5시간인 학생의 점수 예측하기
model.predict([5])
```

# 코딩으로 xor 문제 해결

```python
w11 = np.array([-2,-2])              # 처음 레이어의 처음 가중치
w12 = np.array([2,2])                # 처음 레이어의 두번째 가중치
w2 = np.array([1,1])                 # 두번째 레이어의 가중치
b1=3                                
b2=-1
b3=-1
```

```python
# 퍼셉트론 함수의 정의
def MLP(x,w,b):
    y = np.sum(w*x) + b
    if y <= 0:
        return 0
    else:
        return 1
# NAND 게이트
def NAND(x1,x2):
    return MLP(np.array([x1,x2]),w11,b1)

# or 게이트
def OR(x1,x2):
    return MLP(np.array([x1,x2]),w12,b2)

# AND 게이트
def AND(x1,x2):
    return MLP(np.array([x1,x2]),w2,b3)
# xor 데이트
def XOR(x1,x2):
    return AND(NAND(x1,x2), OR(x1,x2))
```

```python
for x in  [(0,0),(0,1),(1,0),(1,1)]:
    y = XOR(x[0],x[1])
    print(' 입력 값 :', str(x),'출력 값 :', str(y))
```
